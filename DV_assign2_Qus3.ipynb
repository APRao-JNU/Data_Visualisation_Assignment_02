```python
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show, output_notebook
import pandas
```

```python
output_notebook()
```

```python
data = pandas.read_csv('Qus2&3_camera_data_set.csv')
```

```python
source = ColumnDataSource(data = {
    'x': data['Model'],
    'y1': data['Max resolution'],
    'y2': data['Low resolution'],
    'y3': data['Price']
})
```

```python
p1 = figure(width = 300, height = 300, x_range = data['Model'], tools = 'pan, zoom_in, zoom_out, reset', active_drag = 'pan')
p1.circle(x = 'x', y = 'y1', source = source)
```

```python
p2 = figure(width = 300, height = 300, x_range = p1.x_range, y_range = p1.y_range)
"""
p2 = figure(width = 300, height = 300)
p3 = figure(width = 300, height = 300)
p2.x_range = p1.x_range
p2.y_range = p1.y_range
p3.x_range = p1.x_range
p3.y_range = p1.y_range
"""
p2.circle(x = 'x', y = 'y2', source = source)
```

```python
p3 = figure(width = 300, height = 300, x_range = p1.x_range, y_range = p1.y_range)
p3.circle(x = 'x', y = 'y3', source = source)
```

```python
p = gridplot([p1, p2, p3], ncols = 3)
show(p)
```
