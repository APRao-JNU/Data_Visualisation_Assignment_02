```python
import pandas as pd
import numpy as np
import math
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource
from bokeh.layouts import gridplot
from bokeh.io import output_notebook
```

```python
df=pd.read_csv("Qus7_Mall_Customers_dataset.csv")
df
```

```python
# Define the x and y axes for the plot
x = 'Annual Income (k$)'
y = 'Spending Score (1-100)'
```

```python
# Define the rotation angle
angle = np.pi/4
```

```python
# Define the projection
projection = np.array([[np.cos(angle), np.sin(angle)], [-np.sin(angle), np.cos(angle)]])
```

```python
# Apply the projection and rotation to the data
data = np.array([df[x], df[y]])
data = np.dot(projection, data)
df['x_proj'] = data[0]
df['y_proj'] = data[1]
```

```python
# Create a Bokeh ColumnDataSource from the data
source = ColumnDataSource(df)
male_source = ColumnDataSource(df[df["Genre"] == 'Male'])
female_source = ColumnDataSource(df[df["Genre"] == 'Female'])
```

```python
# Define the figure and add a scatter plot
p1 = figure(title='Mall Customer Dataset without projection and rotation', x_axis_label=x, y_axis_label=y)
p1.scatter(x=x, y=y, source=male_source, size=8, fill_alpha=0.5, color='red', legend_label='Male')
p1.scatter(x=x, y=y, source=female_source, size=8, fill_alpha=0.5, color='purple',legend_label='Female')
```

```python
# Define the figure and add a scatter plot
p2 = figure(title='Mall Customer Dataset with projection and rotation', x_axis_label=x, y_axis_label=y)
p2.scatter(x='x_proj', y='y_proj', source=male_source, size=8, fill_alpha=0.5, color='red', legend_label='Male')
p2.scatter(x='x_proj', y='y_proj', source=female_source, size=8, fill_alpha=0.5, color='purple',legend_label='Female')
```

```python
# Add a legend
p1.legend.location = 'top_left'
p1.legend.title = 'Genre'

p2.legend.location = 'top_left'
p2.legend.title = 'Genre'
```

```python
# Create a layout grid for the two plots
plots = gridplot([[p1, p2]])
```

```python
# Displaying the Plot
output_notebook()
```

```python
# Show the plot
show(plots)
```
