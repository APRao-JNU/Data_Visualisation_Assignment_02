```python
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, HoverTool, Circle
from bokeh.plotting import figure, show, output_notebook
import pandas
```

```python
#!pip install bokeh
output_notebook()
```

```python
data = pandas.read_csv('Qus2&3_camera_data_set.csv')
```

```python
source = ColumnDataSource(data = {
    'x': data['Release date'],
    'y': data['Price']
})
```

```python
p = figure(width = 800, height = 300, tools = 'pan, tap, zoom_in, zoom_out, reset', active_drag = None)
cr = p.circle(x = 'x', y = 'y', source = source, fill_color = 'purple', hover_alpha = 0.5, alpha = 0.1, size = 10, selection_color = 'firebrick', nonselection_fill_alpha = 0.1, nonselection_fill_color = 'purple')
```
